```python
x = [1, 2, 3]
y = [3, 4, 5]
x1 = [2, 3, 4]
y1 = [5, 4, 3]
plt.plot(x, y, linewidth= 5)
plt.plot(x1, y1, linewidth = 3)
plt.legend(['first', 'second'])
```

## Stack Plots

```python
days = [1, 2, 3, 4, 5]         # Log of how many days

sleeping = [7, 8, 6, 11, 7]    # How many hours per day spent on sleeping
eating = [2, 3, 4, 3, 2]       # How many hours per day spent on eating
studying = [7, 8, 7, 2, 2]     # How many hours per day spent on studying               
timepass = [8, 5, 7, 8, 13]    # How many hours per day spent on doing tp

plt.stackplot(days, sleeping, eating, studying, timepass, colors=['m', 'c', 'r', 'k'])
```

We cannot assign labels to stackplot. If we want to have the labels anyway, we will have to use empty plots. For example,

```python
plt.plot([], [], color='m', label="Sleeping", linewidth=5)
plt.plot([], [], color='c', label="Eating", linewidth=5)
plt.plot([], [], color='r', label="Studying", linewidth=5)
plt.plot([], [], color='k', label="Timepass", linewidth=5)
plt.stackplot(days, sleeping, eating, studying, timepass, colors=['m', 'c', 'r', 'k'])
plt.legend()
```

## Pie Charts

```python
slices = [100, 20, 200, 35]
names = ["A", "B", "C", "D"]
col = ['m', 'c', 'r', 'y']
plt.pie(slices, labels = names, colors = col)
```

We can also have a custom start angle. For example, if we want to start at 90°, we will do

```python
plt.pie(slices, labels = names, colors = col, startangle = 90)
```

Here is how to add shadow to pie chart

```python
plt.pie(slices, labels = names, colors = col, startangle = 90, shadow = True)
```

If we want to explode the elements (Pop out of the graph), For example if we want to explode B and D -

```python
plt.pie(slices, 
        labels = names, 
        colors = col, 
        startangle = 90, 
        explode = (0, 0.1, 0, 0.2)
       )
```

To add percentages to the pie chart, we do the following

```python
plt.pie(slices, 
        labels = names, 
        colors = col, 
        startangle = 90, 
        explode = (0, 0.1, 0, 0.2),
        autopct = "%1.1f%%"
       )
```

## Getting the Data from Internet (Method 1)

```python
from urllib.request import urlopen
import matplotlib.dates as mdates
```

```python
stock_url = "https://pythonprogramming.net/yahoo_finance_replacement"
source = urlopen(stock_url).read().decode('utf-8')

stock_data = []
split_source = source.split('\n')
```

In our case, the data is clean, but there can be times where the data can have random descriptions or values before the actual data of the stock. In that case, we will filter our the data and append only the numeric values, ignoring the strings. For example, in stocks, we will ignore the labels date, open, close, high, low etc. In our case, there are 7 columns. So, we will do the following

```python
for line in split_source:
    split_line = line.split(",")
    if len(split_line) == 7:
        if 'Open' not in line:
            stock_data.append(line)
```

Now, we load the data of the stock using **np.loadtxt()**. <br>
Here, we call the attribute **converters**, in which we call a dictionary. In that dictionary, key is the index and value is the conversion of the element present in that index of the list to be unpacked. For example, we need to unpack the dates, but the format given in the url is not understood by our machine. So, we need to convert it first. For that, we will define the **convertDate()** function.<br><br>

The arguments to this function is the format of the date. For example, <br>
**%Y** - Full Year (2021) <br>
**%y** - Partial Year (21) <br>
**%m** - Month <br>
**%d** - Day <br>
**%H** - Hours <br>
**%M** - Minutes <br>
**%S** - Seconds

```python
def convertDate(fmt, encoding='utf-8'):
    def bytesConverter(b):
        s = b.decode(encoding)
        return (mdates.datestr2num(s))
    return bytesConverter

date, open_price, high, low, close_price, adjusted_close, volume = np.loadtxt(stock_data, 
                                                                              delimiter=',', 
                                                                              unpack=True,
                                                                              converters = {0: convertDate("%Y%m%d")}
                                                                            )
```

```python
plt.plot_date(date, close_price, "b-", label="Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Stock Price Graph")
plt.legend()
```

## Getting the Data from Internet (Method 2)

```python
from pandas_datareader.data import DataReader
from datetime import datetime
```

```python
# First set start and end date (Range of the dates you want to see data of)
start = datetime(2005, 1, 1)
end = datetime(2021, 1, 1)

# Now, in DataReader, first argument is the stock to trace (TSLA = Tesla)
# Second argument is the source (yahoo = Yahoo Finance API)

# df = DataReader('TSLA', 'yahoo', start, end)
# df = DataReader('TWTR', 'yahoo', start, end)
df = DataReader('TTM', 'yahoo', start, end) # TTM - Tata Motors

#reset_index resets the index of the row just in case it is out of order (while receiving multiple data frames)
df.reset_index(inplace = True)
fig = plt.figure(figsize=(15, 8))
ax = plt.subplot2grid((1, 1), (0, 0))
for label in ax.xaxis.get_ticklabels():
    label.set_rotation(45)
ax.plot_date(df.Date, df.Close, '-', label="Price")
plt.xlabel("Date", fontsize=20)
plt.ylabel("Price", fontsize=20)
plt.plot([], [], label="Profit", linewidth=5, color='g', alpha=0.3)
plt.plot([], [], label="Loss", linewidth=5, color='r', alpha=0.3)
plt.legend(prop={"size": 16})

# Compare the profits or loss of a particular stock considering a base price
# ax.fill_between(df.Date, df.Close, df.Close[0], alpha=0.3)

# See only profit considering a base price
ax.fill_between(df.Date, df.Close, df.Close[0], where=(df.Close > df.Close[0]), facecolor='g', alpha=0.3)
ax.fill_between(df.Date, df.Close, df.Close[0], where=(df.Close < df.Close[0]), facecolor='r', alpha=0.3)

# Now, we can see that there is not a real dividing line, it just fills the color. Now, to plot a horizontal line,
ax.axhline(df.Close[0], color='k', linewidth=1.1)
```

## Basic Customizations

```python
x = np.arange(0, np.pi * 2, 0.05)
y1 = np.sin(x)
y2 = np.cos(x)
```

First, we use plt.figure() to enable the customizations of our graph. Now we use plt.add_axes([x, y, length, width]) function. By using this function, we can get subplots within the same graph. Here, (x, y) is the starting co-ordinate and length, width are the dimensions of the subplot

```python
fig = plt.figure() # Figure is the container for plots. It can store multiple plots inside it

axes1 = fig.add_axes([0, 0, 1, 1]) # start at (0, 0) and have dimension of length = 1 and width = 1
axes2 = fig.add_axes([1.5, 0, 1, 1]) # start at (1.5, 0) and have dimension of length = 1 and width = 1

# Here, we will use axes1.plot() and axes2.plot() instead of plt.plot() since we want to plot on the newly formed axes
axes1.plot(x, y1) 
axes2.plot(x, y2)
```

If we want to make them touch each other, we can just do (1, 0) in the (x, y) part of axes2. Similarly, if we want them to sit on top of each other, we will do (0, 1) in the (x, y) part of axes2

```python
fig = plt.figure() # Figure is the container for plots. It can store multiple plots inside it

axes1 = fig.add_axes([0, 0, 1, 1]) # start at (0, 0) and have dimension of length = 1 and width = 1
axes2 = fig.add_axes([1, 0, 1, 1]) # start at (1, 0) and have dimension of length = 1 and width = 1

# Here, we will use axes1.plot() and axes2.plot() instead of plt.plot() since we want to plot on the newly formed axes
axes1.plot(x, y1) 
axes2.plot(x, y2)
```

There is no restrictionns on the number of axes that can be plotted. For example, if we want a third graph resting on top of first and second, we will do the following:

```python
y3 = np.tan(x)
```

```python
fig = plt.figure() # Figure is the container for plots. It can store multiple plots inside it

axes1 = fig.add_axes([0, 0, 0.5, 0.5]) # start at (0, 0.5) and have dimension of length = 0.5 and width = 0.5
axes2 = fig.add_axes([0.5, 0, 0.5, 0.5]) # start at (0.5, 0) and have dimension of length = 0.5 and width = 0.5
axes3 = fig.add_axes([0, 0.5, 1, 1]) # start at (0, 0.5) and have dimension of length = 1 and width = 1

# Here, we will use axes1.plot() and axes2.plot() instead of plt.plot() since we want to plot on the newly formed axes
axes1.plot(x, y1)
axes2.plot(x, y2)
axes3.plot(x, y3)
```

## Subplots

For plotting the subplots, we need to call the **fig.add_subplot(RCP)** or **fig.add_subplot(R, C, P)** method and within that, we need to call add_subplot() function. In the argument of add_subplot(), we give 3 arguments - R = Row, c = Column, P = Position. For example, if we do plt.figure(221), it means that it will have a 2x2 grid (R = 2 and C = 2), i.e. it will be divided into 4 squares of 1x1, and the plot will take the 1st position (top left)

```python
x = [0, 1, 2, 3, 4]
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x) # plot y = x by taking x as in index array
ax = fig.add_subplot(224)
ax.plot(np.square(x)) # plot y = x^2
```

## Spines and Horizontal Lines
Here, I have used **subplot2grid(shape, start, _rowspan(optional)_ , _colspan(optional)_)**. The first argument takes the shape (how big we want the size to be), and the second argument takes the starting point of the shape.

```python
fig2 = plt.figure()
ax1 = plt.subplot2grid((1, 1), (0, 0))

x = [0, 1, 2, 3, 4]
ax1.plot(x)

# Spines are the borders of matplotlib graph
ax1.spines['left'].set_color('r')
ax1.spines['right'].set_visible(None)
ax1.spines['top'].set_visible(None)
ax1.spines['bottom'].set_linewidth(5)

# Change the color of the tick labels (labels shown on axis)
ax1.tick_params(axis='x', colors='g')

# Draw an horizontal line y = 2:
ax1.axhline(2.5, color='m', linewidth=3)
```

## Live Graphs
The live graphs can be properly shown in pop up window of matplotlib graph, here, since we have to run the cells again and again, it will not be seen properly.

```python
import matplotlib.animation as animation
from matplotlib import style
```

```python
style.use('fivethirtyeight')
fig3 = plt.figure()
ax2 = fig3.add_subplot(111)
graph_data = open("Example.txt", "r").read()
lines = graph_data.split("\n")
x_list = []
y_list = []

# Define function for animation
def animate(interval):
    for line in lines:
        if len(line) > 1:
            x, y = line.split(",")
            x_list.append(float(x))
            y_list.append(float(y))

    ax2.clear() # To clear already existing data
    ax2.plot(x_list, y_list)

# Animate the graph
anim = animation.FuncAnimation(fig3, animate, interval=1000)
```

## Text and Annotations

```python
style.use("default")
fig, ax = plt.subplots()
x = np.arange(0, 4 * np.pi, 0.001)
y = np.sin(x)
ax.plot(x, y)
ax.margins(0)
ax.text(7, 1.5, 'Graph of cos(x)')
ax.annotate("Local Maxima", xy=(np.pi / 2, 1), 
            xytext=(np.pi / 2, 1.8), 
            arrowprops = {'facecolor': 'g', 'shrink': 0.05},
            horizontalalignment='center'
           )

ax.annotate("Local Minima", xy=(3 * np.pi / 2, -1),
            xytext=(3 * np.pi / 2, -1.8),
            arrowprops = dict(arrowstyle = '->', connectionstyle="arc3"),
            horizontalalignment='center'
           )
ax.set_ylim(-2, 2)
```
